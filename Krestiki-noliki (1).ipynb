```python
!pip install pygame
import pygame

# Инициализация Pygame
pygame.init()

# Определение цветов
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)
BLUE = (0, 0, 255)

# Определение размеров окна
WINDOW_SIZE = (300, 300)

# Создание окна
screen = pygame.display.set_mode(WINDOW_SIZE)

# Установка заголовка окна
pygame.display.set_caption("Крестики-нолики")

# Создание игрового поля
board = [[None, None, None], [None, None, None], [None, None, None]]

# Определение текущего игрока
current_player = "X"

# Функция для рисования крестика
def draw_X(row, col):
    pygame.draw.line(screen, BLUE, (col * 100 + 20, row * 100 + 20), (col * 100 + 80, row * 100 + 80), 5)
    pygame.draw.line(screen, BLUE, (col * 100 + 80, row * 100 + 20), (col * 100 + 20, row * 100 + 80), 5)

# Функция для рисования нолика
def draw_O(row, col):
    pygame.draw.circle(screen, BLUE, (col * 100 + 50, row * 100 + 50), 35, 5)

# Функция для проверки выигрыша
def check_win():
    # Проверка строк
    for row in range(3):
        if board[row][0] == board[row][1] == board[row][2] != None:
            return True

    # Проверка столбцов
    for col in range(3):
        if board[0][col] == board[1][col] == board[2][col] != None:
            return True

    # Проверка диагоналей
    if board[0][0] == board[1][1] == board[2][2] != None:
        return True

    if board[0][2] == board[1][1] == board[2][0] != None:
        return True

    # Если никто не выиграл
    return False

# Основной цикл игры
running = True
while running:
    # Обработка событий
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.MOUSEBUTTONDOWN:
            # Получение координат клика мыши
            pos = pygame.mouse.get_pos()
            row = pos[1] // 100
            col = pos[0] // 100

            # Проверка, что клетка свободна
            if board[row][col] == None:
                # Рисование крестика или нолика
                if current_player == "X":
                    draw_X(row, col)
                    board[row][col] = "X"
                    current_player = "O"
                else:
                    draw_O(row, col)
                    board[row][col] = "O"
                    current_player = "X"

                # Проверка на выигрыш
                if check_win():
                    # Вывод сообщения о выигрыше
                    font = pygame.font.Font(None, 36)
                    if current_player == 'X':
                        
                        text = font.render('O' + " Выиграл!", True, BLACK)
                    else:
                        text = font.render('X' + " Выиграл!", True, BLACK)
                    screen.blit(text, (50, 150))
                    pygame.display.flip()

                    # Задержка перед закрытием окна
                    pygame.time.wait(2000)

                    running = False

    # Отрисовка игрового поля
    screen.fill(WHITE)
    for row in range(3):
        for col in range(3):
            if board[row][col] == "X":
                draw_X(row, col)
            elif board[row][col] == "O":
                draw_O(row, col)
    pygame.display.flip()

# Закрытие окна Pygame
pygame.quit()
```
